# Assignment on webScrapping


```python
# iporting libraries
import csv
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# the extract_bird_species checks if the response status code is 200 
#and extracts data from the JSON response if the response is successful
def extract_bird_species(url):
    response = requests.get(url)
    if response.status_code == 200:
        bird_species_data = response.json().get('recordings')

        return bird_species_data
    else:
        print("Failed to fetch the webpage.")
        return None

    # this saves and write data to the species csv file if successful
def save_species_to_csv(data, filename):
    if data:
        with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
            fieldnames = list(data[0].keys())
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            writer.writeheader()
            writer.writerows(data)
        print(f"Species data saved to {filename} successfully.")
    else:
        print("No data to save.")
  

 # this function saves and write data associated with songs 
#to birds_songs csv file if successful
def save_songs_to_csv(data, filename):
    if data:
        songs_data = []
        for recording in data:
            song = {
                'English Name': recording['en'],
                'Bird Song': recording['file']
            }
            songs_data.append(song)
        
        with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
            fieldnames = list(songs_data[0].keys())
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            writer.writeheader()
            writer.writerows(songs_data)
        print(f"Songs data saved to {filename} successfully.")
    else:
        print("No data to save.")

#defining the URL to fetch the bird species and creating files
if __name__ == "__main__":
    url = "https://xeno-canto.org/api/2/recordings?query=cnt:uganda"
    bird_species_data = extract_bird_species(url)
    if bird_species_data:
        
        #saves to bird_species file
        save_species_to_csv(bird_species_data, "bird_species.csv")
    
    #saves to bird_songs file
        save_songs_to_csv(bird_species_data, "bird_songs.csv")
```
