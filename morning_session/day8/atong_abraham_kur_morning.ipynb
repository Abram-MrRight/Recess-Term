```python
# 3. Histogram - Age Distribution: Calculate the age of MPs based on their date of birth and create a histogram to show the distribution of ages. This can provide insights into the age demographics of the MPs.

columns_interested = df2.loc[:, ['Name_of_M.P', 'Date_of_Birth']]
columns_interested
```

```python
# wrong date format
import re

# Define the pattern for valid date format (YYYY-MM-DD)
pattern = r"\d{2}-\d{2}-\d{4}"

# Check for wrong formats in the Date of Birth column
wrong_formats = columns_interested['Date_of_Birth'].apply(lambda x: re.match(pattern, str(x)) is None)

# Filter the DataFrame to show rows with wrong formats
rows_with_wrong_formats = columns_interested[wrong_formats]

# Display the rows with wrong date formats
rows_with_wrong_formats
```

```python
import pandas as pd

# Convert Date of Birth column to datetime
columns_interested['Date_of_Birth'] = pd.to_datetime(columns_interested['Date_of_Birth'], errors='coerce')

# Filter out rows with wrong format dates (year with only two digits)
wrong_format_dates = columns_interested[columns_interested['Date_of_Birth'].dt.year < 1900]

# Correct the wrong format dates by concatenating "19" to the year
columns_interested.loc[wrong_format_dates.index, 'Date_of_Birth'] = "19" + wrong_format_dates['Date_of_Birth'].dt.strftime('%y')

# Remove the time component from the dates
#columns_interested['Date_of_Birth'] = columns_interested['Date_of_Birth'].dt.date

# Display the updated DataFrame
columns_interested.head(20)
```

```python
# Remove rows with NaT (invalid dates)
columns_interested = columns_interested.dropna(subset=['Date_of_Birth'])

# Display the updated DataFrame
columns_interested.head(30)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime


# Convert Date of Birth column to datetime
columns_interested['Date_of_Birth'] = pd.to_datetime(columns_interested['Date_of_Birth'], errors='coerce')

# Calculate age based on Date of Birth
now = datetime.now()
columns_interested['Date_of_Birth'] = now.year - columns_interested['Date_of_Birth'].dt.year

# Convert age to integers
columns_interested['Date_of_Birth'] = columns_interested['Date_of_Birth'].astype(int)

# Plot the histogram
plt.hist(columns_interested['Date_of_Birth'], bins=10, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')

# Display the histogram
plt.show()
```
